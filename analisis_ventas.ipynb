```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer dataset
df = pd.read_csv("datos/sales_sample_2024.csv")

# Mostrar primeras filas
print("Primeras filas del dataset:")
print(df.head())

# =========================
# VENTAS TOTALES
# =========================

ventas_totales = df["sales_amount"].sum()

print("\nVentas Totales:")
print(ventas_totales)

# =========================
# PROMEDIO DE VENTAS
# =========================

promedio_ventas = df["sales_amount"].mean()

print("\nPromedio de Ventas:")
print(promedio_ventas)

# =========================
# VENTAS POR FECHA
# =========================

ventas_por_fecha = df.groupby("sales_date")["sales_amount"].sum()

print("\nVentas por fecha:")
print(ventas_por_fecha)

# =========================
# GRÁFICO
# =========================

plt.figure(figsize=(10,5))

ventas_por_fecha.plot()

plt.title("Evolución de Ventas")
plt.xlabel("Fecha")
plt.ylabel("Monto de ventas")

# Guardar gráfico
plt.savefig("resultados/grafico_ventas.png")

plt.show()

print("\nGráfico guardado en /resultados")
```
